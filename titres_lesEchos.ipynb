```python
import urllib
import bs4
import re
import pandas as pd
from urllib import request
import datetime
import time
import re
import numpy
```

```python
## crÃ©ation d"une liste de dates 
liste_dates=[]
j = datetime.date(2018, 11, 30)
j
```

```python
while j >= datetime.date(2018, 8, 1):
    liste_dates.append(j.strftime("%Y/%m"))
    j = j + datetime.timedelta(days=-1)
liste_dates =set(liste_dates)
liste_dates
```

```python
# liste_dates =["2018/12"]
```

```python
# crÃ©ation de 2 listes l'une contient les articles, l'autre le jour de publication.
liste_titres = []
liste_jour_publication = []
liste_categorie =[]
liste_auteurs =[]
liste_t =[]
df_row = pd.DataFrame()

for d in liste_dates:
    ## exploration des pages internet en ajoutant la date Ã  un dÃ©but d'url
    status = 999
    compt = 1
    while status != 404:
        try:
            page_explorÃ©e = "https://www.lesechos.fr/" + str(d) + "?page=" + str(compt)
            req_1 = urllib.request.Request(page_explorÃ©e,headers={'User-Agent': 'Mozilla/5.0'})
            request_text = request.urlopen(req_1).read()
            page = bs4.BeautifulSoup(request_text, "lxml")
            for url in page.findAll("a" , {'class':"sc-1560xb1-0 fUqzcK sc-1ttlxdz-2 bxQRWI"}):
                url_article = url.get('href', None)
                new_url ="https://www.lesechos.fr"+ url_article
                req_2 = urllib.request.Request(new_url,headers={'User-Agent': 'Mozilla/5.0'})
                request_text2 = request.urlopen(req_2).read()
                page_article = bs4.BeautifulSoup(request_text2, "lxml")
                title=''
                j=''
                a =''
                for auteur in page_article.findAll("span",{"class":"sc-AxirZ sc-1ohdft1-0 bGQROb"}):
                    a = auteur.get_text()
                    liste_auteurs.append(a)
                for ti in page_article.findAll('h1' ,{"class":"sc-AxirZ sc-1ohdft1-0 leQiFa"}):
                    title = ti.get_text()
                    liste_t.append(title)
                for jour_pub in page_article.findAll("span",{'class':"sc-1i0ieo8-0 dBUkqh"}):
                    j = jour_pub.get_text()
                    liste_jour_publication.append(j)
                df_row = df_row.append({'titre': title, 'jour_publication': d , 'auteur':a, 'jour_prÃ©cis':j}, ignore_index=True)
            compt+=1  
        except Exception as e:
            status = 404
len(liste_t)    
```

```python
def nettoie_date(txt):
    text = re.sub(r"PubliÃ© le ", "", txt)
    text = re.sub(r"Mis Ã  jour le ", "", text)
    return text[:-7]
```

```python
df = df_row.copy()
df = df[['titre','jour_publication','auteur','jour_precis']]
df.jour_precis = df.jour_precis.map(lambda x : nettoie_date(x))
df
```
