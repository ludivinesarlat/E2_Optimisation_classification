```python
import urllib
import bs4
import re
import pandas as pd
from urllib import request
import datetime
import time
import re
import codecs
```

```python
## cr√©ation d"une liste de dates 
liste_dates=[]
j = datetime.date(2021, 4, 20)
while j <= datetime.date(2021, 5, 10):
    liste_dates.append(j.strftime("%Y/%m/%d"))
    j = j + datetime.timedelta(days=1)
len(liste_dates)
```

```python
# cr√©ation de 2 listes l'une contient les articles, l'autre le jour de publication.
liste_articles =[]
liste_jour_publication = []
liste_cat1 =[]
liste_cat2 =[]
liste_auteurs =[]
for d in liste_dates:
    ## exploration des pages internet en ajoutant la date √† un d√©but d'url
    page_explor√©e = "https://www.liberation.fr/archives/" + str(d) + "/"
    req_1 = urllib.request.Request(page_explor√©e,headers={'User-Agent': 'Mozilla/5.0'})
    request_text = request.urlopen(req_1).read()
    html = codecs.decode(request_text, 'utf-8')
    page_jour = bs4.BeautifulSoup(html, "lxml")
    compt = 1
    for url in page_jour.findAll("a" , {'class':"color_black decoration_none"}):
        url_article = url.get('href', None)
        print(url_article)
        for titre in url.findAll('h2',{'class':"font_tertiary font_xs font_normal" }):
            liste_articles.append(titre.get_text())
            u = "https://www.liberation.fr" +str(url_article)
            print('u',u)
            req_2 = urllib.request.Request(u,headers={'User-Agent': 'Mozilla/5.0'}) 
            request_text2 = request.urlopen(req_2).read()
            html2 = codecs.decode(request_text2, 'utf-8')
            page_article = bs4.BeautifulSoup(html2, "lxml")
            liste_jour_publication.append(d)
            for cat1 in page_article.findAll("span" , {'class':"color_primary font_primary font_lg font_xl-td"}): 
                liste_cat1.append(cat1.get_text())
            for cat2 in page_article.findAll("a",{"class":"decoration_none margin-xs-right color_grey hover_underline"})[1]:
                liste_cat2.append(cat2)
            for balisehref in page_article.findAll("a"):
                balise = balisehref.get('href', None)
                if balise:
                    if "/auteur/" in balise :
                        auteurs= balise[8:-1]
                else:
                    auteurs="sans auteur"
            liste_auteurs.append(auteurs)
            #print(compt, titre.get_text(), cat1.get_text(), cat2, auteurs)
            compt+=1
```

```python
print(len(liste_articles))
print(len(liste_jour_publication))
print(len(liste_auteurs))
print(len(liste_cat1))
print(len(liste_cat2))
```

```python
del liste_articles[-1]
```

```python
df = pd.DataFrame({ 'titre':liste_articles,'jour_publication':liste_jour_publication,'auteur':liste_auteurs, 'categorie':liste_cat2})
df
```
